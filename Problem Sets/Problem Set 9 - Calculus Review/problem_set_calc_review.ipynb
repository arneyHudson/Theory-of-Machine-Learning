# 4. Newthon's method


$
f' = 15(x - 3)^2\\
$
$
f'' = 30(x-3)
$




```python
func = lambda x: 5*(x-3)**3 + 4
func_prime = lambda x: 15*(x-3)**2
func_double_prime = lambda x: 30*(x-3)

# Newton's method
def newtons_method(func, func_prime, x):
    stop = 10000
    points = [x]
    while stop > 0.001:
        current = x - (func(x) / func_prime(x))
        points.append(current)
        stop = abs(x - current)
        x = current
    return points


initial_guess = -5
points = newtons_method(func, func_prime, initial_guess)

y_changes = [func(point) for point in points]
```

```python
df_points = pd.DataFrame({
    'x': points,
    'y': [func(point) for point in points]
})

df_points.head(len(df_points))
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
x_values = np.linspace(-6, 10, 400)
ax.plot(df_points['x'], df_points['y'], label='$f(x)$')
ax.plot(points, y_changes, 'ro--', label='Newton Method Iterations')

# Annotate the first 5 points and the last point with alternating positions
for i, point in enumerate(points):
    if i < 5 or i == len(points) - 1:
        offset = 0 if i % 2 == 0 else -50
        ax.annotate(f'({point:.2f}, {y_changes[i]:.2f})', (point, y_changes[i]), textcoords="offset points", xytext=(0, offset), ha='center', rotation=45)


make_pretty(ax, 'Newton Method', 'x', 'f(x)')
ax.grid(True)
```
