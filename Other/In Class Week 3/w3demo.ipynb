```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_csv("quad.csv")
df1.head() 
```

```python
class0 = df1.query('label == 0')
class1 = df1.query('label == 1')
plt.scatter(class0["x1"], class0["x2"], color='orange', label="Class 0")
plt.scatter(class1["x1"], class1["x2"], color='blue', label="Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
xmin = df1["x1"].min()
xmax = df1["x1"].max()
#xvals = np.array([xmin, xmax])
xvals = np.linspace(xmin, xmax, 100)
#x2_b1 = 2 * xvals - 8 # This line is linear but the decision boundary should be quadratic
x2_b2 = (-0.2 * (xvals - 2.5)**2 + 3) + 3.5
#plt.plot(xvals, np.array([x2_b1, x2_b2]).T, label=["Boundary 1", "Boundary 2"])
plt.plot(xvals, np.array([x2_b2]).T, label=["Boundary 1"])
plt.scatter(class0["x1"], class0["x2"], color='orange', label="Class 0")
plt.scatter(class1["x1"], class1["x2"], color='blue', label="Class 1")
plt.legend()
```

```python
y_true = np.array([1 if y >= (-0.2 * (x - 2.5)**2 + 3) + 3.5 else 0 for x, y in zip(xvals, xvals)])

# Predicted labels based on the decision boundary
y_pred = np.array([1 if y >= (-0.2 * (x - 2.5)**2 + 3) + 3.5 else 0 for x, y in zip(xvals, xvals)])

# Calculate accuracy
accuracy = accuracy_score(y_true, y_pred)

print(f"Accuracy: {accuracy:.2%}")
```

```python
df2 = pd.read_csv("circle.csv")
df2.sample(n=5).head()
```

```python
class0 = df2.query('label == 0')
class1 = df2.query('label == 1')
plt.scatter(class0["x1"], class0["x2"], color='orange', label="Class 0")
plt.scatter(class1["x1"], class1["x2"], color='blue', label="Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
#df2["x3"] = df2["x1"] + df2["x2"] # What is a good derived feature?
df2["x3"] = np.sqrt(df2["x1"]**2 + df2["x2"]**2)
df2.sample(n=5).head()
```

```python
class0 = df2.query('label == 0')
class1 = df2.query('label == 1')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(class0["x1"], class0["x2"], class0["x3"], color='orange', label="Class 0")
ax.scatter(class1["x1"], class1["x2"], class1["x3"], color='blue', label="Class 1")

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

plt.show()
```
